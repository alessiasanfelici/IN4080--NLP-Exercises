```python
import nltk
nltk.download("brown")
```

```python
from nltk.corpus import brown
wordlist = brown.tagged_words()
wordlist[:20]
```

```python
# Create a dictionary to store the counts
d = {}
```

```python
import collections
```

```python
for word, tag in wordlist:
    # Simplify the tags
    tag = tag.split("-")[0].split("+")[0]

    # Create a dictionary for each tag
    if tag not in d:
        d[tag] = collections.defaultdict(int)

    # Fill the tag dictionary with len and count
    word_length = len(word)
    d[tag][word_length] += 1  
```

```python
d
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(d)
df = df.fillna(0)
df = df.sort_index()
df
```

Which are the five most frequent word lengths?

```python
d2 = {}
for index, row in df.iterrows():
    d2[index] = row[1:].sum()
print(d2)
```

```python
sorted_d2 = dict(sorted(d2.items(), key=lambda item: item[1], reverse = True))
print("Five most freqent word lengths:")
for i in range(5):
    print(list(sorted_d2.items())[i])
```

Which are the five most frequent part-of-speech tags?

```python
d1 = {}
for col in df.columns:
    d1[col] = df[col].sum()
sorted_d1 = dict(sorted(d1.items(), key=lambda item: item[1], reverse = True))
print("Five most frequent part-of-speech tags:")
for i in range(5):
    print(list(sorted_d1.items())[i])
```

Create two additional dataframes with the conditional probability distributions. One of them should 
represent ð‘ƒ(tag = ð‘¡ | length = ð‘™) and the other one ð‘ƒ(length = ð‘™ | tag = ð‘¡).

```python
df_conditional_tag = df.copy()
for tag in df.columns:
    df_conditional_tag[tag] = df[tag]/df[tag].sum()
df_conditional_tag
```

```python
df_conditional_length = df.copy()
for length in df.index:
    df_conditional_length.loc[length] = df.loc[length]/df.loc[length].sum()
df_conditional_length
```

What is the probability that a common noun (NN) has length 5?

```python
df_conditional_length.loc[5, "NN"]
```

What is the probability that a word of length 1 is an article (AT)?

```python
df_conditional_tag.loc[1, "AT"]
```

What is the average length of adjectives (JJ)?

```python
sum(df.index * df.loc[:, "JJ"])/sum(df.loc[:, "JJ"])
```

```python
from nltk.corpus import movie_reviews
```

```python
movie_reviews.categories()
movie_reviews.fileids("pos")
movie_reviews.raw('pos/cv000_29590.txt')
```

```python
movie_docs = [(movie_reviews.raw(fileid), label) for label in movie_reviews.categories() for fileid in movie_reviews.fileids(label)]
movie_docs[:5]
```

```python
df = pd.DataFrame(movie_docs , columns=['text', 'label'])
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df["text"], df["label"], test_size=0.1)
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
x_train_bow = cv.fit_transform(x_train)
```

```python
cv.vocabulary_
cv.vocabulary_["boring"]
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(x_train_bow, y_train)
```

```python
x_val_bow = cv.transform(x_val)
predicted_y_val = nb.predict(x_val_bow)
```

```python
from sklearn import metrics
acc = metrics.accuracy_score(y_val, predicted_y_val)
acc
```

```python
print(metrics.classification_report(y_val, predicted_y_val))
```

```python
print(metrics.confusion_matrix(y_val, predicted_y_val))
```
